### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# The pyplot interface and the object oriented interface. The pyplot is simpler and faster and is recommended to new users and simple tasks. 
# The object oriented interface is more complex and more complete and is recommended to more advanced and professional applications.
# 
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figures: The page where the plot appears, it can contain multiple plots.
# Axes: The plot area where the data is plotted. Each plot has at least (and usually) 2 axes but can also contain tree axes in a 3d plot.
# Axis: Each axes has two or three axis. X and Y, and Z in 3d plots. 
# All the visuals. Like lines, labels and text. 
```

### When plotting in Matplotlib, what is the expected input data type?

```python
# Lists, Arrays, Panda series or data frames, Tuples. 
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import matplotlib.pyplot as plt
import numpy as np

# The pyplot approach
x = np.linspace(-4, 4)
plt.figure(figsize=(6,4))
plt.plot(x, x**2, label='Quadratic')
plt.xlabel('Test')
plt.ylabel('Tset')
plt.title('Plot with test and tset')
plt.legend()
plt.show()

# The object oriented approach
```

```python
import matplotlib.pyplot as plt
import numpy as np

# The object oriented approach
x = np.linspace(-4, 4)
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x, x**2, label='Quadratic*')
ax.set_xlabel('Test')
ax.set_ylabel('Tset')
ax.set_title('Plot with test and tset')
ax.legend()

```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

fig, (ax, ax2) = plt.subplots(1, 2, figsize=(12,4))
ax.scatter(random_data_x, random_data_y, label='Random numbers')
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_title('Scatterplot of random numbers')
ax.legend()

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())


ax2.bar(names, values)
ax2.set_xlabel('Fruit')
ax2.set_ylabel('Value')
ax2.set_title('Bar of fruit')


```
